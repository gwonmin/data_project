```python
df = pd.read_excel('C:/Users/didrn/OneDrive/바탕 화면/1review.xlsx', engine = 'openpyxl')
```

```python
text = ""

for i in df["리뷰상세내용"]:
    text += i
```

```python
token_ko = t.nouns(text)

ko = nltk.Text(token_ko)
```
